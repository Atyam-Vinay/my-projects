```python
import pandas as pd #importing necessary libraries
import numpy as np
train_data=pd.read_csv("/content/sample_data/mnist_train_small.csv")
test_data=pd.read_csv("/content/sample_data/mnist_test.csv") #importing the locally saved mnist data
train_data=pd.DataFrame(train_data)
test_data=pd.DataFrame(test_data)#specifying train and test data
train_data
```

```python
test_data
```

```python
train_data.rename(columns={'6':'target'},inplace=True)
test_data.rename(columns={'7':'target'},inplace=True)#renaming the columns
y_train=np.asarray(train_data['target'])
y_test=np.asarray(test_data['target']) #dividing feature and target data
X_train=np.asarray(train_data.drop('target',axis=1))
X_test=np.asarray(test_data.drop('target',axis=1))
```

```python
X_train.shape,y_train.shape
```

```python
import matplotlib.pyplot as plt
one_image=X_train[30]
plt.imshow(one_image.reshape(28,28))
```

```python
X_train=X_train/255 #scaling the train data
X_test=X_test/255 #scaling the test data
```

```python
from keras.models import Sequential
from keras.layers import Dense #importing keras librraries
model=Sequential([Dense(10,input_shape=(784,),activation='sigmoid')])
model #creating model object
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy']) #building the model
```

```python
model.fit(X_train,y_train,epochs=10)#training the model
```

```python
model.evaluate(X_test,y_test) #model evaluation
```

```python
plt.imshow(X_test[5].reshape(28,28)) #taking sample from test data
```

```python
y_predicted=model.predict(X_test)
np.argmax(y_predicted[5]) #prediction using the model
```
